```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    mytextblob = TextBlob(text)
    mysentiment = mytextblob.sentiment
    return mysentiment
```

```python
get_sentiment("Perfect for kids, adults and everyone")
```

```python
dataset['verified_reviews'].dropna().apply(get_sentiment)
```

```python
def get_sentiment_polarity(text):
    mytextblob = TextBlob(text)
    mysentimentpolarity = mytextblob.sentiment.polarity
    return mysentimentpolarity
```

```python
dataset["polarity"] = dataset['verified_reviews'].dropna().apply(get_sentiment_polarity)
```

```python
dataset
```

```python
def get_sentiment_subjectivity(text):
    mytextblob = TextBlob(text)
    mysentimentsubjectivity = mytextblob.sentiment.subjectivity
    return mysentimentsubjectivity
```

```python
dataset["subjectivity"] = dataset['verified_reviews'].dropna().apply(get_sentiment_subjectivity)
```
